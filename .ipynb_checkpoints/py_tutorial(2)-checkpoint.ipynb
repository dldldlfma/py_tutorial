## 4. fucntion(함수)

함수는 반복되는 작업을 하나로 묶어서 편하게 처리하기 위해 사용합니다. 

대체로 **입력에 대한 출력이 정형화 되어 있는 경우**에 사용하는 것이 아주 좋습니다.  


### 4.1 함수의 정의
python에서 함수는 아래와 같이 정의 합니다. 


```python
def 함수이름(입력값):
    수행할 일
    return 출력값
```

우리는 학교에서 함수를 아래와 같이 배웠습니다. 

f(x) = 2x^2 +5

위 함수를 python 함수로 나타내면 어떻게 될까요?

```python
def func(x):
    fx = 2*x**2 +5
    return fx
```



````python
def func(x):
    fx = 2*x**2 +5
    return fx
````

````python
func(5)
````

### 4.2 함수에 관한 추가적 이야기

함수에 입력값이 없으면 어떻게 될까요?  

혹은 함수에 출력값이 없다면 어떻게 될까요?  

상관습니다.

입력값이 없다면 

```python
def func():
    return ~~
```

위 코드 처럼 입력값을 안받으면 되고  

출력값이 없다면

```python
def func(a,b,c):
    return
```

리턴값을 안주면 됩니다.  

입력값이 몇개 인지 모르지만 다받아야 된다면 어떻게 해야될까요?  

```python
def func(*args):
    sum =0
    for i in args:
        sum = sum + i
    return sum
```

위와 같이 하면 됩니다. 


### 함수 안에서 변수의 유효범위

함수안에서 사용가능한 변수는 함수로 입력받은 변수와 함수 안에서 선언된 변수 뿐입니다.  

함수안에서 외부 변수를 사용하려면 어떻게 해야될까요?  

global 이라는 명령어를 사용하면 됩니다. 

```python
g_val = 0
def global_value():
    global g_val
    g_val = g_val+10

global_value()
print(g_val)
```

# 5. Class

* 객체지향 프로그래밍의 시작
* 재활용 가능한 코드작성이 편해짐
* 어렵게 생각하면 한없이 어렵다 하지만 꼭 그렇지만은 않다.
* 진짜 어려운건 남들이 짜놓은 코드가 어려운거지 class가 어려운게 아니다 
* 꼭 필요한것만 배우고 모르는게 나오면 찾아서 해결하자.

## 5.1 Class의 생성

```python
class Music():
    def __init__(self,name,composer,lyricist,singer,album_name):
        self.name = name #곡명
        self.composer = composer #작곡가
        self.lyricist = lyricist #작사가
        self.singer = singer #가수
        self.album_name = album_name
        self.ranking = None
        self.last_ranking = None
            
    def set_ranking(self, ranking):
        self.last_ranking = self.ranking
        self.ranking = ranking

        
```

class는 self를 통해서 자신이 가지고 있는 모든 내용물을 전달합니다. 

__init__(self, ~~ )함수는  self를 통해 자기 자신이 가지고 있는 내용을 포함하여 다양한 내용을 입력으로 받습니다.   
__init__(self, ~~ )함수는 class를 생성하면 바로 실행됩니다.   
__init__(self, ~~ )함수 안에서 변수를 선언하면 해당 class안에 포함되는 변수가 됩니다.   

위 내용 처럼 직접 class를 선언해보면서 class를 어떻게 선언하는지를 익혀봅시다.

아래와 같은 class를 선언해보면서 python class의 사용법을 익혀 보겠습니다. 

* 직원
* 학교
* 카페



## 5.2 Class의 사용

몇가지 class를 쓰기위해서 알아둬야할 점들이 있다. 

* self.value = value 왼쪽과 같은 방식으로 내부 변수를 선언한다. 
* self.`__`value = value 왼쪽과 같이 .`__`xxx 형태로 선언하면 **private 변수**로 외부에서 직접적인 접근이 불가능하다.
* 함수에서 내부 변수를 포함하여 사용하고 싶다면 class내부 모든함수는 선언시 첫번째 인자 값으로 self를 받는다. 
* 부모클래스로부터 상속받는 개념을 이해한다.


## 5.3 Class 상속

기존 class를 상속받는 형태로 class를 생성할 수도 있다.

아래 **Employee class**는 **Person class**를 상속받는다. 
 
```python
class Person:
    def __init__(self,name,age):
        self.name =name
        self.age =age
    def print_info(self):
        print(self.name, self.age)

class Employee(Person):
    def __init__(self,name=None,age=None,selary=None,Number='x'):
        super().__init__(name,age)
        self.No = Number
        self.selary = selary
    
    def print_info2(self):
        print(self.name, self.age,self.selary,self.No)


if __name__ == "__main__":
    #a = Employee('ksg',29,4000,5)
    a = Employee('ksg')
    a.print_info()
    a.print_info2()
```

## 5.4 연산자 오버로딩


연산자 오버로딩을 통해 기존에 python에서 제공하는 연산 기능을 class에서 사용할 수 있도록 해줄수 있다.

`+ - * / 와 같은 연산들을 사용할수 있도록 해주는 방법중 하나라고 생각하면 좋다.`


```python
class Person:
    def __init__(self,name,age):
        self.name =name
        self.age =age
    def print_info(self):
        print(self.name, self.age)

class Employee(Person):
    def __init__(self,name=None,age=None,selary=None,Number='x'):
        super().__init__(name,age)
        self.No = Number
        self.selary = selary
    
    def print_info2(self):
        print(self.name, self.age,self.selary,self.No)

    def __add__(self, value):
        print(self.name,value)

if __name__ == "__main__":
    #a = Employee('ksg',29,4000,5)
    a = Employee('ksg')
    a.print_info()
    a.print_info2()
    a+"hello"
```
